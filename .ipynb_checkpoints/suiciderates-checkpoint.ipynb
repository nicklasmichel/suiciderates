# Suicide Rates

This analysis uses the WHO data set __Suicide rate estimates__ which compares socio-economic info with suicide rates by year and country. http://apps.who.int/gho/data/node.main.MHSUICIDEASDR?lang=en

Let's start by importing the prepared xlsx-file and take a first look at the data.

```python
import pygal
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from pygal.style import DefaultStyle
from IPython.display import SVG, display
%matplotlib inline 
%pylab inline


# note that the xlsx file needs to be in the same location as your notebook
df = pd.read_excel('who_data.xlsx')
# prints information about the data frame including column dtypes and non-null values
df.info()
# return the first 5 rows of the data frame
df.head()
```

Only missing value appear in column 'code' which is later used for plotting purposes. The package 'pygal' does not support all countries.

__1. Where do the most/least suicides take place?__

```python
# subset data frame accordingly
data = df[(df['year'] == 2016) & (df['sex'] == 'Both sexes')][['code', 'suicides_per_100k']]
# sort values by suicide rates
data.sort_values(by='suicides_per_100k', ascending=False, inplace=True)
# drop numerical index
data.reset_index(drop=True, inplace=True)
# set 'code' column as new index
data.set_index('code', inplace=True)
# drop missing country codes
data.dropna(inplace=True)
# store the results in two data frames to distinguish between low and high suicide rates
data_split = np.array_split(data, 2)

# initialize plot of world map
worldmap_chart = pygal.maps.world.World(style=DefaultStyle, show_legend=False)
# add data to plot
worldmap_chart.add('Low', data_split[1].to_dict()['suicides_per_100k'])
worldmap_chart.add('High', data_split[0].to_dict()['suicides_per_100k'])
# display plot as support vector graphic
display(SVG(worldmap_chart.render(disable_xml_declaration=True)))
```

```python
# subset data frame accordingly
data = df[(df['year'] == 2016) & (df['sex'] == 'Both sexes')][['country', 'suicides_per_100k']]
# sort values by suicide rates
data.sort_values(by='suicides_per_100k', ascending=False, inplace=True)
# drop numerical index
data.reset_index(drop=True, inplace=True)
# set 'code' column as new index
data.set_index('country', inplace=True)
# drop missing country codes
data.dropna(inplace=True)
# get first/last n entries
n = 30
plot = pd.concat([data.head(n), data.tail(n)])

# create the plot with correct size, colors, labels
plt.rcParams["figure.figsize"] = (20, 20)
barplot = sns.barplot(x=plot['suicides_per_100k'], y=plot.index, orient='h', palette='bwr')
barplot.set_ylabel('Country',fontsize=20)
barplot.set_xlabel('Suicides per 100k Population', fontsize=20)
barplot.tick_params(labelsize=15)
```

```python
# subset data frame accordingly
d = df[(df['year'] == 2016) & (df['sex'] == 'Both sexes')]
# desired column names of new table
cols = ['Suicides per 100k', 'Least Suicidal Country', 'Most Suicidal Country']
# initialize data frame
tbl = pd.DataFrame(index=list(d['region'].unique()), columns=cols)
# loop over every region
for region in d['region'].unique():
    # calculate average suicide rate per region
    tbl.loc[region, 'Suicides per 100k'] = round(
        d[d['region'] == region]['suicides_per_100k'].mean(), 1)
    # find least suicidal country per region
    lowest_country = d[(d['region'] == region) & (d['suicides_per_100k'] == d[d['region'] == region]
    ['suicides_per_100k'].min())]['country'].iloc[0]
    # and its actual suicide rate
    lowest_suicide = d[(d['region'] == region) & (d['suicides_per_100k'] == d[d['region'] == region]
    ['suicides_per_100k'].min())]['suicides_per_100k'].iloc[0]
    # combine information and write it in data frame
    lowest_country_suicide = str(lowest_country) + ' (' + str(lowest_suicide) + ')'
    tbl.loc[region, 'Least Suicidal Country'] = lowest_country_suicide
    # same procedure for most suicidal country
    highest_country = d[(d['region'] == region) & (d['suicides_per_100k'] == d[d['region'] == region]
    ['suicides_per_100k'].max())]['country'].iloc[0]
    highest_suicide = d[(d['region'] == region) & (d['suicides_per_100k'] == d[d['region'] == region]
    ['suicides_per_100k'].max())]['suicides_per_100k'].iloc[0]
    highest_country_suicide = str(highest_country) + ' (' + str(highest_suicide) + ')'
    tbl.loc[region, 'Most Suicidal Country'] = highest_country_suicide
# print resulting table
tbl
```

__2. Are men more suicidal than women?__

Let's start by comparing the yearly, worldwide suicide rates for both sexes.

```python
# calculate the yearly, average suicide rate for both sexes
tbl2 = pd.DataFrame(df.groupby(['year']).mean()['suicides_per_100k'])
# rename column
tbl2.rename({'suicides_per_100k': 'Both Sexes'}, axis='columns', inplace=True)
# calculate the yearly, average suicide rate for men
tbl2['Male'] = df[df['sex']=='Male'].groupby(['year']).mean()['suicides_per_100k']
# calculate the yearly, average suicide rate for women
tbl2['Female'] = df[df['sex']=='Female'].groupby(['year']).mean()['suicides_per_100k']
tbl2['Ratio (%)'] = round(tbl2['Male'] / (tbl2['Male'] + tbl2['Female']) * 100, 1)
tbl2.round(2)
```

Figure 3 lists the top 10 countries with the lowest/highest suicide ratio among men.

```python
# subset data frame accordingly
d = df[df['year'] == 2016]
countries = []
ratios = []

# loop over all countires
for country in d['country'].unique():
    # store country in countries list
    countries.append(country)
    # store male suicide rate
    male = d[(d['country'] == country) & (d['sex'] == 'Male')]['suicides_per_100k'].values
    # store female suicide rate
    female = d[(d['country'] == country) & (d['sex'] == 'Female')]['suicides_per_100k'].values
    # calculate male suicide ratio
    ratio = male / (female + male)  
    # store ratio in ratios list
    ratios.append(ratio)

# flatten ratios array and multiply by 100 and round
ratios = [round(x, 4) * 100 for x in np.concatenate(ratios).ravel().tolist()]
# create data frame 
data = pd.DataFrame(data=ratios, index=countries, columns=['Ratio (%)'])
# sort values by suicide ratios
data.sort_values(by='Ratio (%)', ascending=False, inplace=True)
# get first/last n entries
n = 10
plot = pd.concat([data.head(n), data.tail(n)])[:-1]

# create the plot with correct size, colors, labels
plt.rcParams["figure.figsize"] = (20, 8)
barplot = sns.barplot(x=plot['Ratio (%)'], y=plot.index, orient='h', palette='PiYG')
barplot.set_ylabel('Country',fontsize=25)
barplot.set_xlabel('Male suicide ratio (%)', fontsize=25)
barplot.tick_params(labelsize=20)
```

```python
# create box plot: men vs. women suicide rates
box_plot = pygal.Box(box_mode="tukey", range=(0, 50))
box_plot.add('Male', df[df['sex'] == 'Male']['suicides_per_100k'].values)
box_plot.add('Female', df[df['sex'] == 'Female']['suicides_per_100k'].values)

# display plot as support vector graphic
display(SVG(box_plot.render(disable_xml_declaration=True)))
```

__3. Is a country's suicide rate related to its size, income or HDI?__

Create correlation matrix.

```python
# create dictionary which maps categorical income variable to integer based variable to be able to calculate correlations
inc_dic = {'Low income': 1, 'Lower middle income': 2, 'Upper middle income': 3, 'High income': 4}
df['Income'] = [inc_dic[x] for x in df['income']]
corr = df.corr().round(4)
names = {'suicides_per_100k': 'Suicides', 'land_area': 'Land Area', 'hdi': 'HDI Factor'}
corr.rename(index=names, columns=names, inplace=True)
corr * 100
```

Create box plot for suicide rate per income category.

```python
# create box plot: men vs. women suicide rates
box_plot = pygal.Box(box_mode="tukey", range=(0, 30))
box_plot.add('High', df[df['income'] == 'High income']['suicides_per_100k'].values)
box_plot.add('Upper middle', df[df['income'] == 'Upper middle']['suicides_per_100k'].values)
box_plot.add('Lower middle', df[df['income'] == 'Lower middle income']['suicides_per_100k'].values)
box_plot.add('Low', df[df['income'] == 'Low income']['suicides_per_100k'].values)

# display plot as support vector graphic
display(SVG(box_plot.render(disable_xml_declaration=True)))
```

Create scatter plot to show linear dependency between hdi factor and suicide rates.

```python
d = df.dropna()
plt.scatter(d['hdi'], d['suicides_per_100k'], color='black')
linfit = np.polyfit(d['hdi'].values, d['suicides_per_100k'].values, 1)
plt.plot([0.3, 1], [linfit[1], linfit[1] + linfit[0]], lw=5)
plt.gca().set_ylim(0, 25)
plt.xlabel('HDI factor', fontsize=18)
plt.ylabel('Suicides per 100k', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
```
